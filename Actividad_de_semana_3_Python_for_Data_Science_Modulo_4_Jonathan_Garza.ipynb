Leyendo Archivos con Open

```python
#Importar librería urllib.request
import urllib.request

#Definir URL de donde descargar el archivo muestra
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/example1.txt'

#Definir nombre del archivo a descargar
filename = 'example1.txt'

#Definir objeto
urllib.request.urlretrieve(url, filename)

#Descargar Archivo
!wget -O example1.txt https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/example1.txt

#Definir nombre del archivo a abrir
example1 = "example1.txt"

#Abrir archivo en modo de lectura
file1 = open(example1, "r")

#Desplegar nombre del archivo
print('El nombre del archivo es:',file1.name)

#Deplegar el modo en quese abrió el archivo r: lectura o w: escritura
print('El modo en el que se abrió es:', file1.mode)

#Leer el contenido del archivo
FileContent = file1.read()
print('El contenido del archivo es:\n'+FileContent)

#Desplegar el tipo de datos en el archivo:
print('El tipo de datos del archivo es:',type(FileContent))

#Siempre se debe de cerrar un archivo cuando se termine de trabajar con él:
file1.close()

#Una mejor forma de abrir un archivo es usar 'with', se abre y se cierra automáticamente el archivo aunque existan excepciones durante el proceso
with open(example1, "r") as file1:
    #Leer el contenido del archivo
    FileContent = file1.read()
    print('El contenido del archivo es:\n'+FileContent)

#Verificar si realmente se cerró el archivo al terminar 'with'
print('¿El archivo se cerró?:',file1.closed)

#Abrir archivo e imprimir los primeros cuatro caracteres
with open(example1, "r") as file1:
    print('Los primeros cuatro caracteres del archivo son:',file1.read(4))

#Abrir archivo e imprimir caracteres en secciones
with open(example1, "r") as file1:
    print('Los primeros cuatro caracteres del archivo son:',file1.read(4))
    print('Los siguientes cuatro caracteres del archivo son:',file1.read(4))
    print('Los siguientes siete caracteres del archivo son:',file1.read(7))
    print('Los siguientes quince caracteres del archivo son:',file1.read(15))

#Leer e imprimir la primera línea del archivo
with open(example1, "r") as file1:
    print("La primera línea del archivo es:",file1.readline())

#Leer número específico de caracteres en una línea
with open(example1, "r") as file1:
    print("Los primeros 20 caracteres de la primera línea del archivo son:",file1.readline())
    print('Los siguientes veinte caracteres del archivo son:',file1.read(20))

#También se pueden usar ciclos for para iterear entre las líneas
with open(example1,"r") as file1:
        i = 0;
        for line in file1:
            print("Iteración", str(i), ": ", line)
            i = i + 1

#Abrir el archivo y leer todas la líneas y asignarlas a una lista
with open(example1, "r") as file1:
    FileasList = file1.readlines()

print('El contenido del archivo en formato lista es el siguiente:\n',FileasList)
```

Escribiendo Archivos con Open

```python
#Definir el nombre del archivo
exmp2 = 'Example2.txt'

#Abrir el archivo en modo escritura y escribir una línea
with open(exmp2, 'w') as writefile:
    writefile.write("This is line A")

# Abrir y leer archivo para verficar su contenido
with open(exmp2, 'r') as testwritefile:
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

#Abrir el archivo en modo escritura y escribir dos líneas
with open(exmp2, 'w') as writefile:
    writefile.write("This is line A\n")
    writefile.write("This is line B\n")

#Abrir y leer archivo para verficar su contenido
with open(exmp2, 'r') as testwritefile:
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

#Definir una lista con saltos de línea
Lines = ["This is line A\n", "This is line B\n", "This is line C\n"]
print('El contenido de la lista es el siguiente:\n',Lines)

#Escribir el contenido de la lista en en un archivo
with open('Example2.txt', 'w') as writefile:
    for line in Lines:
        print(line)
        writefile.write(line)

#Abrir y leer archivo para verficar su contenido
with open(exmp2, 'r') as testwritefile:
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

#Utilizar el modo 'w'o de escritura, sobreescribe los contenidos del archivo
with open('Example2.txt', 'w') as writefile:
    writefile.write("Overwrite\n")

#Abrir y leer archivo para verficar su contenido
with open(exmp2, 'r') as testwritefile:
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

#Abrir el texto en modo 'a' o añadir, no sobreescribe el contenido actual, sólo añade los datos nuevos
with open('Example2.txt', 'a') as testwritefile:
    testwritefile.write("This is line C\n")
    testwritefile.write("This is line D\n")
    testwritefile.write("This is line E\n")

#Abrir y leer archivo para verficar su contenido
with open(exmp2, 'r') as testwritefile:
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

#Modos adicionales para abrir archivos

#r+ : Lectura y Escritura. No se puede truncar el archivo.
#w+ : Escritura y Lectura. Trunca el archivo.
#a+ : Añadir y Lectura. Crea un archivo si no existe ninguno.

#Abrir el texto en modo 'a+' o añadir y lectura, escribir datos y luego leerlos
with open('Example2.txt', 'a+') as testwritefile:
    testwritefile.write("This is line E\n")
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

#No se desplego nada porque el puntero en este modo no regresa al punto inicial del archivo.

#Comandos para el caso anterior
#.tell() - regresa la posición actual del puntero en bytes
#.seek(offset,from) - cambia la posición del puntero en 'offset' con respecto a la poscición 'from'. 'from' puede tomar los valores de 0,1,2 correspondientes al principio, la posición actual y el final.

with open('Example2.txt', 'a+') as testwritefile:
    print('Posición inicial:', testwritefile.tell())
    
    data = testwritefile.read()
    if (not data):  #Los strings vacíos regresan falso en python
            print('No hay datos') 
    else: 
            print(testwritefile.read())
            
    testwritefile.seek(0,0) # moverse 0 bytes desde el principio o regresar al principio.
    
    print("\nNueva ubicación : {}".format(testwritefile.tell()))
    data = testwritefile.read()
    if (not data): 
            print('No hay datos') 
    else: 
            print(data)
    
    print("Ubicación después de la lectura: {}".format(testwritefile.tell()) )

#Ambos métodos w+ y r+ tienen acceso a los métodos de lectura y escritura, la diferencia es que w+ sobreescribe todos los datos exixstentes en el archivo.
#Para trabajar con un archivo con datos existentes, la opción es usar r+ y a+. Al usar r+, puede ser utilizar el método .truncate() al final de los datos.
#Esto reducirá el tamaño de los datos y borrará todo lo demás.

#Ejemplo sin usar .truncate()
with open('Example2.txt', 'r+') as testwritefile:
    data = testwritefile.readlines()
    testwritefile.seek(0,0) #iniciar al principio del archivo
   
    testwritefile.write("Línea 1" + "\n")
    testwritefile.write("Línea 2" + "\n")
    testwritefile.write("Línea 3" + "\n")
    testwritefile.write("finalizado\n")
    #testwritefile.truncate()
    testwritefile.seek(0,0)
    print(testwritefile.read())

#Ejemplo utilizando .truncate()
with open('Example2.txt', 'r+') as testwritefile:
    data = testwritefile.readlines()
    testwritefile.seek(0,0) #iniciar al principio del archivo
   
    testwritefile.write("Línea 1" + "\n")
    testwritefile.write("Línea 2" + "\n")
    testwritefile.write("Línea 3" + "\n")
    testwritefile.write("finalizado\n")
    testwritefile.truncate()
    testwritefile.seek(0,0)
    print(testwritefile.read())

#Copiar archivos

#Leer un archivo y copiarlo a otro
with open('Example2.txt','r') as readfile:
    with open('Example3.txt','w') as writefile:
          for line in readfile:
                writefile.write(line)

#Abrir y leer archivo para verficar su contenido
with open('Example3.txt', 'r') as testwritefile:
    print('El contenido del archivo es el siguiente:\n',testwritefile.read())

```

Leer, Escribir y Procesar Datos con Pandas

```python
#Importar la librería Pandas y asignarla a pd
import pandas as pd

#Importar librería urllib.request
import urllib.request

#Definir un dataframe
df=pd.DataFrame({'a':[1,2,1],'b':[1,1,1]})

print('Datos Originales:\n{}'.format(df))

#Encontrar los valores únicos en la columna 'a'
print('Datos de a con valores únicos:\n{}'.format(df['a'].unique()))

#Obtener como resultado un dataframe que sólo contenga valores menores a 2
print('Datos de a con menores a 2:\n{}'.format(df[df['a']<2]))

#Definir URL de donde descargar el archivo muestra
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/TopSellingAlbums.csv"

#Definir nombre del archivo a descargar
filename = 'TopSellingAlbums.csv'

#Definir objeto
urllib.request.urlretrieve(url, filename)

#Descargar Archivo CSV
!wget -O TopSellingAlbums.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/TopSellingAlbums.csv

#Crear un dataframe de Pandas con el archivo que se descargó
df = pd.read_csv("TopSellingAlbums.csv")

#Desplegar las primeras 5 filas de datos
print('Datos de archivo CSV:\n{}'.format(df.head()))

#Definir URL de donde descargar el archivo muestra
xlsx_path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%204/Datasets/TopSellingAlbums.xlsx'

#Definir nombre del archivo a descargar
filename = 'TopSellingAlbums.xlsx'

#Definir objeto
urllib.request.urlretrieve(xlsx_path, filename)

#Descargar Archivo xlsx
!wget -O TopSellingAlbums.xlsx https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%204/Datasets/TopSellingAlbums.xlsx

#Crear un dataframe de Pandas con el archivo que se descargó
df = pd.read_excel("TopSellingAlbums.xlsx")

#Desplegar las primeras 5 filas de datos
print('Datos de archivo Excel:\n{}'.format(df.head()))

#Acceder a la columna lenght, obtener un dataframe y desplegar los datos
print('Dataframe de Lenght del archivo Excel:\n{}'.format(df[['Length']]))

#Acceder a la columna lenght, obtener una serie y desplegar los datos
print('Serie de Lenght del archivo Excel:\n{}'.format(df['Length']))

#Acceder a varias columnas, obtener un dataframe y desplegar los datos
print('Dataframe de varias columnas del archivo Excel:\n{}'.format(df[['Artist','Length','Genre']]))

#Acceder al valor en la primer fila y primer columna y desplegar los datos
print('Valor en la primer fila y primer columna del archivo Excel:\n{}'.format(df.iloc[0, 0]))

#Acceder al valor en la primer fila y tercer columna y desplegar los datos
print('Valor en la primer fila y tercer columna del archivo Excel:\n{}'.format(df.iloc[0,2]))

#Acceder al valor en la segunda fila y tercer columna y desplegar los datos
print('Valor en la segunda fila y tercer columna del archivo Excel:\n{}'.format(df.iloc[1,2]))

#Acceder al valor en la primer fila en la columna artista y desplegar los datos
print('Valor en la primer fila en la columna artista del archivo Excel:\n{}'.format(df.loc[0, 'Artist']))

#Acceder al valor en la segunda fila en la columna artista y desplegar los datos
print('Valor en la segunda fila en la columna artista del archivo Excel:\n{}'.format(df.loc[1, 'Artist']))

#Acceder al valor en la primer fila en la columna publicado y desplegar los datos
print('Valor en la primer fila en la columna publicado del archivo Excel:\n{}'.format(df.loc[0, 'Released']))

#Acceder al valor en la segunda fila en la columna publicado y desplegar los datos
print('Valor en la segunda fila en la columna publicado del archivo Excel:\n{}'.format(df.loc[1, 'Released']))

#Aplicar slicing con iloc en el dataframe y desplegar los datos
print('Valores de la primera a la segunda fila de la primera a la tercera columna del archivo Excel:\n{}'.format(df.iloc[0:2, 0:3]))

#Aplicar slicing con loc en el dataframe y desplegar los datos
print('Valores de la primera a la tercera fila de la columna artista a la columna publicado del archivo Excel:\n{}'.format(df.loc[0:2, 'Artist':'Released']))

#Convertir los índices del dataframe a letras por medio de una lista
new_index=['a','b','c','d','e','f','g','h']
df_new=df
df_new.index=new_index
print('Valores en a,Artist del archivo Excel:\n{}'.format(df_new.loc['a', 'Artist']))
print('Valores en a:d,Artist del archivo Excel:\n{}'.format(df_new.loc['a':'d', 'Artist']))
```
