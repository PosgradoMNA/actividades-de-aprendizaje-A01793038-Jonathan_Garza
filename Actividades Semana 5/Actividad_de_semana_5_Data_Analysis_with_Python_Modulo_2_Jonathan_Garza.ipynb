```python
#Importar librerías para los ejercicios
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

**Se descarga y crea la base de datos**

```python
#Establecer la url de donde se descargará el archivo csv para los ejercicios
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

#Se define una lista con el nombre deseado de las columnas
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

#Descargar el archivo, asignarlo a un dataframe y asignar el nombre deseado de las columnas.
df = pd.read_csv(path, names = headers)

#Visualizar el encabezado del archivo
df.head()
```

```python
# Se buscan todos los valores iguales a "?" en todo el dataframe y se convierten a NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Si se desean encontrar los valores no numéricos se utiliza la función .isnull() si se desean encontrar los valores numéricos se usa .notnull()

#Se crea un nuevo dataframe con el resultado de la búsqueda de valores no numéricos
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Se puede utilizar un ciclo for para contar los valores faltantes en todas las columnas

#Se transforman las columnas del dataframe en una lista y se realiza el ciclo for
for column in missing_data.columns.values.tolist():
    #Se imprime la columna actual
    print(column)
    #Se imprime la cuenta de valores faltantes
    print (missing_data[column].value_counts())
    #Se imprime un espacio
    print("")
```

**Se comienza con la limpieza del dataframe**

```python
#Los valores faltantes de la columna "normalized-losses" se reemplazarán por el promedio

#Se transforman los valores de la columna a float y se calcula el promedio de la columna
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)

#Se imprime el valor del promedio
print("Promedio de normalized-losses:", avg_norm_loss)

#Se bucan las posiciones vacías (las que tienen una condición True) y se reemplazan por el promedio
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

#Los valores faltantes de la columna "bore" se reemplazarán por el promedio

#Se transforman los valores de la columna a float y se calcula el promedio de la columna
avg_bore=df['bore'].astype('float').mean(axis=0)

#Se imprime el valor del promedio
print("Promedio de bore:", avg_bore)

#Se bucan las posiciones vacías (las que tienen una condición True) y se reemplazan por el promedio
df["bore"].replace(np.nan, avg_bore, inplace=True)

#Se transforman los valores de la columna a float y se calcula el promedio de la columna
avg_stroke = df["stroke"].astype("float").mean(axis = 0)

#Se imprime el valor del promedio
print("Promedio de stroke:", avg_stroke)

#Se bucan las posiciones vacías (las que tienen una condición True) y se reemplazan por el promedio
df["stroke"].replace(np.nan, avg_stroke, inplace = True)

#Se transforman los valores de la columna a float y se calcula el promedio de la columna
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)

#Se imprime el valor del promedio
print("Promedio de horsepower:", avg_horsepower)

#Se bucan las posiciones vacías (las que tienen una condición True) y se reemplazan por el promedio
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

#Se transforman los valores de la columna a float y se calcula el promedio de la columna
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)

#Se imprime el valor del promedio
print("Promedio de peak rpm:", avg_peakrpm)

#Se bucan las posiciones vacías (las que tienen una condición True) y se reemplazan por el promedio
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)

#Se puede usar la función .value_counts() para contar los valores que están presentes en una columna
print("La cuenta de los valores de ''num-of-doors' es:\n{}".format(df['num-of-doors'].value_counts()))

#Se obtiene el valor más frecuente
valor_mas_frecuente = df['num-of-doors'].value_counts().idxmax()

#Se reemplazan los valores faltantes de 'num-of-doors' con el valor más frecuente
df["num-of-doors"].replace(np.nan, valor_mas_frecuente, inplace=True)

#Se quitan las columnas que no tienen los valores en la columna precio ya que esta es la variable dependiente
df.dropna(subset=["price"], axis=0, inplace=True)

#Se reestablecen los índices ya que se quitaron dos columnas
df.reset_index(drop=True, inplace=True)

#Ya no se deberían tener valores faltantes, se procede a verificar
print("Valores faltantes por columna:")
print(df.isnull().sum()) 

#Se imprime el encabezado del dataframe sin valores faltantes
df.head()
```

**Se procede a verifica que el tipo de datos por columna es correcto**

```python
#La función .dtypes analiza todo el dataframe y te devuelve el tipo de datos presente en cada columna
print('Los tipos de dato presentes en cada columna son:\n\n{}'.format(df.dtypes)+'\n')

#Se convierten los datos a los tipos que deberían de tener:

#Tipo float
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")

#Tipo int
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")

print('Los nuevos tipos de dato presentes en cada columna son:\n\n{}'.format(df.dtypes)+'\n')
```

**Estandarización de Datos**

```python
#Se pueden realizar operaciones matemáticas de forma directa con los datos en Pandas

#Convertir los datos en MPG a L/100km
df["city-mpg"] = 235/df["city-mpg"]
df["highway-mpg"] = 235/df["highway-mpg"]

#Renombrar las columnas para reflejar estos cambios
df.rename(columns={'city-mpg': 'city-L/100km','highway-mpg':'highway-L/100km'}, inplace=True)

#Checar los datos posterior a las transformaciones
df.head()
```

**Normalización de Datos**

```python
#Normalización de datos es transformar las magnitudes de los mismos a rangos en los que sea más fácil
#compararlos en caso de que las magnitudes cambien mucho entre las columnas. Típicamente se busca tener valores
#entre 0 y 1.

#Reemplazar los valores originales con la fórmula valor original/máximo valor original
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()

#Desplegar el encabezado de las columnas cambiadas para visualizar los resultados
df[['length','width','height']].head()
```

**Agrupación de Datos**

```python
#Se transforma la columna que contenga los valores en el formato requerido
df["horsepower"]=df["horsepower"].astype(int, copy=True)

#Se importa la librería requerida para realizar las gráficas
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

#Cambiar las etiquetas de x/y y el nombre de gráfica para que contengan los datos correctos
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Se desean realizar 4 agrupaciones entre los valores máximo y mínimo. Se utiliza numpy linspace, para realizar la lista
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)

#Se despliegan los grupos
print('Los nombres de los grupos de potencia son:\n\n{}'.format(bins)+'\n')

#Se definen los nombres de los grupos:
group_names = ['Low', 'Medium', 'High']

#Se utiliza la función .cut() para determinar a qué grupo pertenece cada valor en la columna
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )

#Se cuentan y se despliegan los valores para saber cuántos elementos contiene cada grupo
print('La cantidad de elementos en cada grupo de potencia es:\n\n{}'.format(df["horsepower-binned"].value_counts())+'\n')

#Se despliega encabezado para visualizar los resultados
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Se grafica la distribución de cada grupo:
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

#Cambiar las etiquetas de x/y y el nombre de gráfica para que contengan los datos correctos
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#En vez de realizar todas las operaciones de forma separada, la gráfica histograma se utiliza para realizar lo anterior
plt.pyplot.hist(df["horsepower"])

#Cambiar las etiquetas de x/y y el nombre de gráfica para que contengan los datos correctos
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

**Variables Indicadoras**

```python
#Estas variables sustituyen por números las variables categóricas con el objetivo que los métodos de regresión puedan
#realizar las operaciones con ellas ya que sólo funcionan con valores numéricos

#Obtener las variables indicadoras de tipo de combustible
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#Se reemplazan los nombres de la columnas para tener calridad de que son las mismas que se usaron en la columna original
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
#Se añaden las nuevas columnas que se crearon 
df = pd.concat([df, dummy_variable_1], axis=1)

#Se eliminan las columnas originales ya que no son requeridas
df.drop("fuel-type", axis = 1, inplace=True)

#Se despliega el encabezado para visualizar cambios
df.head()
```

```python
# Se realiza la misma operación pero para el tipo de aspiración y se asigna a "dummy_variable_2"
dummy_variable_2 = pd.get_dummies(df['aspiration'])

#Se eliminan las columnas originales ya que no son requeridas
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

#Se despliega el encabezado para visualizar cambios
dummy_variable_2.head()
```

```python
#Se añaden las nuevas columnas que se crearon
df = pd.concat([df, dummy_variable_2], axis=1)

#Se eliminan las columnas originales ya que no son requeridas
df.drop('aspiration', axis = 1, inplace=True)

#Se despliega el encabezado para visualizar cambios
df.head()
```
